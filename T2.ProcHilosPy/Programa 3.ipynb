```python
import time

def TareaHijo():
    print("Proceso Hijo con PID : {}".format(mp.current_process().pid))
    time.sleep(3)
    print("Fin del proceso hijo")

def main():
    print("Proceso Padre con PID : {}".format(mp.current_process().pid))
    miproceso = mp.Process(target=TareaHijo)
    miproceso.start()
    miproceso.join()
    print("Fin del proceso padre")    

main()
```

```python
def TareaHijo(x):
    print("PID padre: {}".format(x))
    print("Proceso Hijo con PID : {}".format(mp.current_process().pid))
    print("Proceso con nombre: {}". format(mp.current_process.name))
    print("Fin del proceso hijo")



def main():
    names = ["Proceso 1, Proceso 2, Proceso 3"]
    procesos = []
    for i in range(len(names)):
        procesos.append(mp.Process(target=TareaHijo, name=names[i], args=(mp.current_process().pid,)))

    for i in range(len(procesos)):
        procesos[i].start()

    for i in range(len(procesos)):
        procesos[i].join()


main()

```

```python
def TareaProceso():
    pactual = mp.current_process()
    print("Proceso hijo : {}".format(pactual.pid))
    time.sleep(5)
    print("continua la ejecucion")

miproceso = mp.Process(target=TareaProceso)
miproceso.start()
time.sleep(10)
print("Proceso padre ha terminado, el proceso main")
print("Proceso hijo terminando...")
miproceso.terminate()
print("Proceso hijo finaliza con exito")
```

```python
from  multiprocessing import Process, Queue

def worker(num,q):
    print("Se pone en la cola : ", num*num)
    q.put(num*num)

if __name__=="__main__":
    my_queue = Queue()
    p = Process(target=worker, args = (5, my_queue))
    p.start()
    p.join()

    print("Se lee de la cola: ", my_queue.get())
```
