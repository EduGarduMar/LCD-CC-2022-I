```python
import multiprocessing as mp

def tarea(cadena):
  print("Hola", cadena)

if __name__ == "__main__":
  p = mp.Process(target=tarea, args=("Lucas C.",))
  p.start()
  p.join()
```

```python
def calc_cuad(numeros):
  print("Calcula el cuadrado de los numeros")
  for i in numeros:
    print("Cuadrado:", i*i)


nums = range(10)
p1 = mp.Process(target=calc_cuad, args=(nums,))
p1.start()
p1.join()

print("Termina la ejecucion numérica")
```

```python
def calc_cub(numeros):
  print("Calcula el cuadrado de los numeros")
  for i in numeros:
    print("El cubo es", i**3)

nums = range(10)
p2 = mp.Process(target=calc_cub, args=(nums,))
p2.start()
p2.join()
```

```python
import time 

nums = range(10)
t = time.time()
calc_cuad(nums)
calc_cub(nums)

print("Tiempo de ejecución: ", time.time()-t)
print("Termina la ejecución")
```

```python
t = time.time()
p1 = mp.Process(target = calc_cuad, args=(nums,))
p2 = mp.Process(target=calc_cub, args=(nums,))

p1.start()
p2.start()

p1.join()
p2.join()

print("Tiempo de ejecución: ", time.time()-t)
print("Termina la ejecución")
```
